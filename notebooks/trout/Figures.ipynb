```python
def get_limits(nb):
    top = nb + 2 * math.sqrt(nb / 2)
    bottom = nb - 2 * math.sqrt(nb / 2)
    return top, bottom

ratios = []
nbs = [10, 20, 30, 40, 50, 75, 100, 200]
for nb in nbs:
    top, bottom = get_limits(nb)
    print(nb, top)
    ratios.append(top / nb)
fig, ax = plt.subplots()
ax.plot(nbs, ratios, '.-')
```
